```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LinearRegression
```

```python
energy_data = pd.read_csv("energydata_complete .csv")
energy_data
```

```python
energy_data.info()
```

```python
energy_data.describe()
```

```python
#Question 12
lm = LinearRegression()
lm.fit(energy_data[["T2"]],energy_data[["T6"]])

# calculating the R^2 value in two d.p;
round(r2_score(energy_data[["T6"]], lm.predict(energy_data[["T2"]])), 2)
```

```python
#Question 13

energy_data_one = energy_data.drop(labels = ['date', 'lights'], axis = 1)

scaler = MinMaxScaler()
energy_data_one = pd.DataFrame(data = scaler.fit_transform (energy_data_one), columns = energy_data_one.columns, index = energy_data_one.index)

# setting features
X = energy_data_one.drop(labels = ['Appliances'], axis = 1)
y = energy_data_one.loc[:, 'Appliances']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

lm = LinearRegression()
lm.fit(X_train, y_train)

predictions = lm.predict(X_test)
print("mae:", round(mean_absolute_error(y_test, predictions), 2))
```

```python
#Question 14
print("rss:", round(np.sum(np.square(y_test - predictions)), 2))
```

```python
#Question 15
print("rmse:", round(mean_squared_error(y_test, predictions, squared = False), 3))
```

```python
#Question 16
print("r2:", round(r2_score(y_test, predictions), 2))
```

```python
#Question 17
weights = pd.DataFrame(data = lm.coef_, columns = ["weights"], index = lm.feature_names_in_)
weights.sort_values(by = "weights")
```

```python
#Question 18
ridge = Ridge(alpha = 0.4)
ridge.fit(X_train, y_train)

predictions = ridge.predict(X_test)
print("rmse:", mean_squared_error(y_test, predictions, squared = False))
```

```python
#Question 19 and 20
lasso = Lasso(alpha = 0.001)
lasso.fit(X_train, y_train)

preds = lasso.predict(X_test)
print("rmse:", round(mean_squared_error(y_test, preds, squared = False), 3))
lasso.coef_
```
